# Exemplo básico

Este notebook mostra três formas de usar o pacote `ebird2cbro`:

1. usando uma **checklist/lista pessoal** do eBird;
2. usando um **hotspot** do eBird;
3. usando uma **planilha exportada diretamente do eBird**.

A função adiciona registros do eBird como colunas de presença/ausência na lista do CBRO.

Na saída:

- `1` = espécie registrada;
- `0` = espécie não registrada;
- células vazias = linhas que não são espécies, por exemplo ordem, família ou gênero.

Primeiro, instale o pacote com o link do github:

Depois disso, importe a função principal:

```python
from ebird2cbro import ebird2cbro
```

Agora vamos rodar a função `ebird2cbro()`.

Primeiro, para um checklist/lista pessoal. Use source_type="checklist" e coloque em "source" o código da checklist. Se você já estiver trabalhando com uma planilha do CBRO e quiser adicionar novas colunas na sua lista, basta informar o caminho/nome com "cbro_path".

```python
df_checklist = ebird2cbro(
    source="S123456789",                     # substitua pelo código real da checklist
    source_type="checklist",
    column_name="minha_checklist",          # aqui vai o nome da coluna que será adicionada na lista de espécies do CBRO
    api_key="SUA_CHAVE_DE_API_EBIRD",       # substitua pela sua chave de API do eBird
    output_path="CBRO_com_checklist.xlsx"   # aqui vai o caminho/nome do arquivo de saída
)
```

Agora, para um hotspot. Use source_type="hotspot" e coloque em "source" o código do hotspot. Novamente, se você já estiver trabalhando com uma planilha do CBRO e quiser adicionar novas colunas na sua lista, basta informar o caminho/nome com "cbro_path".

```python
df_hotspot = ebird2cbro(
    source="L123456",                       # substitua pelo código real do hotspot
    source_type="hotspot",
    column_name="meu_hotspot",              # aqui vai o nome da coluna que será adicionada na lista de espécies do CBRO    
    api_key="SUA_CHAVE_DE_API_EBIRD",       # substitua pela sua chave de API do eBird
    output_path="CBRO_com_hotspot.xlsx"     # aqui vai o caminho/nome do arquivo de saída
)
```

Agora, usando uma planilha exportada do eBird. Neste caso, a função adiciona **uma coluna para cada checklist** presente na planilha. Aqui você não precisa informar api_key nem column_name. Aqui também, se você já estiver trabalhando com uma planilha do CBRO e quiser adicionar novas colunas na sua lista, basta informar o caminho/nome com "cbro_path".

```python
df_export = ebird2cbro(
    source="ebird_export.xlsx",              # planilha exportada do eBird
    source_type="file",
    output_path="CBRO_com_export_ebird.xlsx" # aqui vai o caminho/nome do arquivo de saída
)
```

# Basic Example

This notebook demonstrates a simple example of how to use `ebird2cbro`.

The goal is to retrieve a species list from eBird (either from a personal checklist or a hotspot), match it with the official CBRO checklist, and create a new presence/absence column indicating which species were recorded.

First, import the main function.

```python
from ebird2cbro import ebird2cbro
import os
```

The next step is load the eBird API key

For examples using eBird checklists or hotspots, an API key is required. We recommend storing it as an environment variable instead of writing it directly in the code.

```python
api_key = os.getenv("EBIRD_API_KEY")
```

Now we run the main function `ebird2cbro()`

The package will:
- retrieve species from eBird,
- match them against the CBRO checklist,
- and add a new presence/absence column

```python
df = ebird2cbro(
    cbro_path="Lista_CBRO_2021.xlsx",    # here you should put your CBRO species list file
    source="S123456789",                 # here you should put your eBird checklist code
    source_type="checklist",             # here you should put your source type (checklist or region)
    column_name="Dom_21_jul_2024",       # this is the column name in your CBRO species list file
    api_key=api_key,                     # here you should put your eBird API key
    output_path="CBRO_com_ebird.xlsx"    # this is the output file path (and name)
)
```

Now we inspect the resulting dataframe

It contains the original CBRO checklist plus the new column with `1`, species recorded in the eBird list and `0`, species not recorded.

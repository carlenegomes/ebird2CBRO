# Basic Example

This notebook demonstrates a simple example of how to use `ebird2cbro`.

The goal is to retrieve a species list from eBird (either from a personal checklist or a hotspot), match it with the official CBRO checklist, and create a new presence/absence column indicating which species were recorded.

1. Import the package

```python
from ebird2cbro import ebird2cbro
import os
```

2. Load the eBird API key 

```python
api_key = os.getenv("EBIRD_API_KEY")
```

3. Run `ebird2cbro()` using an eBird checklist  

```python
df = ebird2cbro(
    cbro_path="Lista_CBRO_2021.xlsx",
    source="S260534489",
    source_type="checklist",
    column_name="Dom_21_jul_2024",
    api_key=api_key,
    output_path="CBRO_com_ebird.xlsx"
)
```

4. Inspect the resulting table  

```python
df.head()
```

The final output is a CBRO spreadsheet with an additional column containing:

- `1` → species recorded in the eBird list  
- `0` → species not recorded
